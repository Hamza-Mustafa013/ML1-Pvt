```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'car_img.jpg'
img = cv2.imread(path)
plt.imshow(img)
plt.show()
```

```python
img1 = cv2.rectangle(img, pt1 = (500,250), pt2 = (150,400), color = (255,0,0), thickness = 3)
plt.imshow(img1)
plt.show()
```

```python
path = 'small_cars.jpg'
img2 = cv2.imread(path)
plt.imshow(img2)
plt.show()
```

```python
img3 = cv2.rectangle(img2, pt1 = (150,1200), pt2 = (1500,2100), color = (255,0,0), thickness = 15)
plt.imshow(img3)
plt.show()
```

```python
img4 = cv2.rectangle(img3, pt1 = (1600,1200), pt2 = (3000,2100), color = (255,0,0), thickness = 15)
plt.imshow(img4)
plt.show()
```
