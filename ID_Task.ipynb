```python
import numpy as np

path = "ID_Card.jpg"
img = cv2.imread(path)

bright_factor = 1.2
brightened_img = np.clip(img * bright_factor, 0,255).astype(np.uint8)

plt.imshow(brightened_img)
plt.show()
```

```python
# Read the original image
original_image = cv2.imread('ID_Card.jpg')

# define rotation range
angles = range(0,360,1)

# Output directory
output_directory = 'C:/Users/Hamza_Mustafa/ML1-Bootcamp/rotated_images/'

# iterations through angles

for angle in angles:
#     make copy of original image
    rotated_img = np.array(original_image)
    rows,cols,shp = rotated_img.shape
    rotation_matrix = cv2.getRotationMatrix2D((rows/2,cols/2),angle,1)
    rotated_img = cv2.warpAffine(rotated_img,rotation_matrix,(cols,rows))
    
    # Save the rotated image with a filename indicating the angle
    output_filename = f'{output_directory}rotated_image_{angle}.jpg'
    cv2.imwrite(output_filename, rotated_image)

print('Rotation completed.')
```
