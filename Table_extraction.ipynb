```python
import fitz  # PyMuPDF

def search_page_by_heading(pdf_path, heading_text):
    doc = fitz.open(pdf_path)

    for page_number in range(doc.page_count):
        page = doc[page_number]
        text = page.get_text()
        
        if heading_text in text:
            print(f'Heading found on page {page_number + 1}')
            return page_number + 1  # Page numbers start from 1

    print('Heading not found in the PDF')
    return None

# Example usage
pdf_path = 'meta.pdf'
heading_to_search = 'CONSOLIDATED BALANCE SHEETS'

page_number = search_page_by_heading(pdf_path, heading_to_search)

if page_number is not None:
    print(f'The page number containing the heading is: {page_number}')
```

```python
import fitz  # PyMuPDF

def search_pages_by_heading(pdf_path, heading_text):
    doc = fitz.open(pdf_path)
    matching_pages = []

    for page_number in range(doc.page_count):
        page = doc[page_number]
        text = page.get_text()

        if heading_text in text:
            matching_pages.append(page_number + 1)  # Page numbers start from 1

    if not matching_pages:
        print('Heading not found in the PDF')
        return None
    else:
        print(f'Heading found on pages: {matching_pages}')
        return matching_pages

# Example usage
pdf_path = 'meta.pdf'
heading_to_search = 'CONSOLIDATED BALANCE SHEETS'

pages_with_heading = search_pages_by_heading(pdf_path, heading_to_search)

if pages_with_heading is not None:
    print(f'The pages containing the heading are: {pages_with_heading}')
```

```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path, start_page, end_page):
    doc = fitz.open(pdf_path)

    text = ''
    for page_number in range(start_page - 1, min(end_page, doc.page_count)):
        page = doc[page_number]
        text += page.get_text()

    doc.close()
    return text

pdf_path = 'meta.pdf'
start_page = 80
end_page = 85

result_text = extract_text_from_pdf(pdf_path, start_page, end_page)

print(result_text)
```

```python
import openpyxl

# Load the Excel workbook
workbook = openpyxl.load_workbook('Book1.xlsx')

# Assuming Meta and Template sheets are present
meta_sheet = workbook['Meta']
template_sheet = workbook['Template']

# Get the value in A5 cell from Meta sheet
search_word = meta_sheet['A5'].value
search_target = search_word.split()
last_value = search_target[-1]
# Find the index of the search_target in modified_list
index = [i for i, value in enumerate(modified_list) if value == last_value]

# Get the next two values from modified_list
result_values = modified_list[index[0] + 1:index[0]+3]
# print(result_values)

# Get the value in B5 cell from Meta sheet
search_value_B5 = meta_sheet['B5'].value

# Find the cell containing the search_value_B5 in the Template sheet
search_cell = None

for row_number, row in enumerate(template_sheet.iter_rows(min_row=1, max_col=1, values_only=True), start=1):
    if row[0] == search_value_B5:
        search_cell = template_sheet.cell(row=row_number, column=2)  # Assuming B5 corresponds to column 2
        break

# If a match is found, update B6 and C6 cells
if search_cell:
    template_sheet.cell(row=search_cell.row + 1, column=search_cell.column).value = result_values[0]
    template_sheet.cell(row=search_cell.row + 1, column=search_cell.column + 1).value = result_values[1]

# Save the changes to the workbook
workbook.save('Book1.xlsx')
```
