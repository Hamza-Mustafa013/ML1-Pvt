```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'abc.png'
img = cv2.imread(path)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
img_rgb.shape
```

```python
lower_red = np.array([100,0,0])
upper_red = np.array([255,50,50])
red_mask = cv2.inRange(img_rgb,lower_red,upper_red)

black_color = (0,0,0)
img_rgb[np.where(red_mask)] = black_color
```

```python
plt.imshow(img_rgb)
plt.show()
```
