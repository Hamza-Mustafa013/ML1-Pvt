```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Trading-features.csv")
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
# Create a new column 'Price_Status' and initialize it with 'No'
df['Price_Status'] = 0
```

```python
# Check if the current price is greater than the previous row's price
df.loc[df['last_price'] > df['last_price'].shift(1), 'Price_Status'] = 1
```

```python
df.head()
```

```python
df = df.drop(['created_at','id'],axis=1)
df
```

```python
class_count = df['Price_Status'].value_counts()
class_count
```

```python
X = df.iloc[:,:-1]
X
```

```python
y = df.iloc[:,-1:]
y
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=43)
```

```python
# Initialize the SVM model
from sklearn.svm import SVC
classifier = SVC()
```

```python
# Train the model
classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = classifier.predict(X_test)
```

```python
# Evaluate the mode
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# accuracy = accuracy_score(y_test, y_pred)
accuracy = classifier.score(X_test,y_test)
print(f"Accuracy: {accuracy:.2f}")
```

```python
class_report = classification_report(y_test, y_pred)
print(f"Classification Report:\n{class_report}")
```
