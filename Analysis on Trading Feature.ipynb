```python
# Histogram of 'last_price'
plt.figure(figsize=(10,6))
sns.histplot(data['last_price'],kde=True,color='Blue')
plt.xlabel('Last Price')
plt.ylabel('Frequency')
plt.title('Distribution of Last Price')
plt.show()
```

```python
# line plot of 'date' against 'last_price'
data['created_at'] = pd.to_datetime(data['created_at'])
plt.figure(figsize=(15, 10))
sns.lineplot(x=data['created_at'], y=data['last_price'])
plt.xlabel('date')
plt.ylabel('Last Price')
plt.title('Line Plot of date against Last Price')
plt.show()
```

```python
# Box plot of 'last_price' to identify outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['last_price'])
plt.ylabel('Last Price')
plt.title('Box Plot of Last Price')
plt.show()
```
